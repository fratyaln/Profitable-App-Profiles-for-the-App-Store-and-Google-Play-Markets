```python
def non_english(df):
    non_english_app={}
    for index ,i in df.items():  
        non_asci=0
        for char in i:
          if ord(char) >127:
                non_asci+=1
        if non_asci>3:
             non_english_app[i]=index
    return non_english_app
```

```python
non_english_app=non_english(Playstore_df["App"])
Playstore_df = Playstore_df.drop(non_english_app.values()).reset_index(drop=True)
```

```python
non_english_app=non_english(Appstore_df["track_name"])
Appstore_df = Appstore_df.drop(non_english_app.values()).reset_index(drop=True)
```

```python
def non_free_apps(df):
    non_free_app_index=[]
    for index ,i in df.items():  
       if i>0.0:
        non_free_app_index.append(index)
    return non_free_app_index
```

```python
Playstore_df[Playstore_df["Type"]!="Free"]
```

```python
non_free_app_index=non_free_apps(Appstore_df["price"])
Appstore_df=Appstore_df.drop(non_free_app_index).reset_index(drop=True)
```

```python
Playstore_df['Price'] = pd.to_numeric(Playstore_df['Price'], errors='coerce')
```

```python
non_free_app_index=non_free_apps(Playstore_df["Price"])
Playstore_df=Playstore_df.drop(non_free_app_index).reset_index(drop=True)
```

```python

def freq_table_percentages(df, column_name):
    table = df[column_name].value_counts(normalize=True) * 100
    return table

def display_freq_table(df, column_name):
    table = freq_table_percentages(df, column_name)
    table = table.sort_values(ascending=False)
    for index, value in table.items():
        print(f"{index}: {value:.2f}%")
```

```python
display_freq_table(Appstore_df, 'prime_genre')
```

```python
display_freq_table(Playstore_df, 'Category')
```
